```python
# load_data 函数（带异常处理）
def load_data(filepath="china_data.json"):
    """
    读取JSON文件，返回字典格式：{省份: {'capital':省会, 'abbreviation':简称}}
    若文件不存在或读取失败，则使用内置默认数据
    """
    try:
        with open(filepath, 'r', encoding='utf-8') as f:
            data_list = json.load(f)
        # 转换为字典
        data_dict = {}
        for item in data_list:
            province = item['province']
            data_dict[province] = {
                'capital': item['capital'],
                'abbreviation': item['abbreviation']
            }
        print(f"✓ 成功加载数据文件：{filepath}，共 {len(data_dict)} 个省级行政区")
        return data_dict
    except FileNotFoundError:
        print(f"⚠ 文件 {filepath} 不存在，使用内置默认数据")
        return _get_builtin_data()
    except Exception as e:
        print(f"✗ 读取文件出错：{e}，使用内置默认数据")
        return _get_builtin_data()
```

```python
# get_capital 函数
def get_capital(province, data):
    """根据省份名返回省会，不存在返回 None"""
    if province in data:
        return data[province]['capital']
    else:
        return None
```

```python
# get_abbreviation 函数
def get_abbreviation(province, data):
    """根据省份名返回简称"""
    if province in data:
        return data[province]['abbreviation']
    else:
        return None
```

```python
# get_province_by_capital 函数
def get_province_by_capital(capital, data):
    """根据省会反查省份名（忽略大小写和首尾空格）"""
    capital_clean = capital.strip().lower()
    for province, info in data.items():
        if info['capital'].strip().lower() == capital_clean:
            return province
    return None
```

```python
# get_random_province 函数
def get_random_province(data):
    """随机返回一个省份名"""
    import random  # 确保在函数内导入，避免依赖单元格顺序
    provinces = list(data.keys())
    return random.choice(provinces)
```

```python
def is_answer_correct(user_input, correct_answer):
    """
    智能比较器：判断用户输入是否与正确答案匹配。
    包含：去空格、忽略大小写、以及针对省份/省会的常见别名容错。
    """
    if not user_input or not correct_answer:
        return False
    
    user_clean = user_input.strip().lower()
    correct_clean = correct_answer.strip().lower()
    
    # 1. 精确匹配（忽略大小写和空格）
    if user_clean == correct_clean:
        return True
    
    # 2. 别名容错（例如："北京" 匹配 "北京市"）
    # 如果正确答案是 "北京市"，用户输入 "北京" 也算对
    if correct_clean.endswith('市') and user_clean == correct_clean[:-1]:
        return True
    if correct_clean.endswith('自治区') and user_clean == correct_clean[:-3]:
        return True
    if correct_clean.endswith('特别行政区') and user_clean == correct_clean[:-5]:
        return True
    
    # 3. 反向容错：用户输入全称，答案可能是简称（如 答案"内蒙古"，用户输入"内蒙古自治区"）
    if user_clean.endswith('市') and correct_clean == user_clean[:-1]:
        return True
    if user_clean.endswith('自治区') and correct_clean == user_clean[:-3]:
        return True
    if user_clean.endswith('特别行政区') and correct_clean == user_clean[:-5]:
        return True
    
    return False
```

```python
def ask_capital(province, data):
    """问省会模式"""
    correct = data[province]['capital']
    user_input = input(f"请回答 '{province}' 的省会城市是？\n> ").strip()
    is_correct = is_answer_correct(user_input, correct)
    
    # 即时反馈
    if is_correct:
        print("✅ 回答正确！")
    else:
        print(f"❌ 回答错误。正确答案是：{correct}")
    return is_correct, correct

def ask_abbreviation(province, data):
    """问简称模式"""
    correct = data[province]['abbreviation']
    user_input = input(f"请回答 '{province}' 的简称是？\n> ").strip()
    is_correct = is_answer_correct(user_input, correct)
    
    if is_correct:
        print("✅ 回答正确！")
    else:
        print(f"❌ 回答错误。正确答案是：{correct}")
    return is_correct, correct

def ask_province(capital, data):
    """根据省会问省份模式"""
    # 利用A的查询函数反查省份
    correct = get_province_by_capital(capital, data)
    if not correct:
        print(f"⚠ 系统错误：未找到省会 '{capital}' 对应的省份。")
        return False, None
    
    user_input = input(f"请回答 '{capital}' 是哪个省的省会？\n> ").strip()
    is_correct = is_answer_correct(user_input, correct)
    
    if is_correct:
        print("✅ 回答正确！")
    else:
        print(f"❌ 回答错误。正确答案是：{correct}")
    return is_correct, correct
```

```python
# ============================================
# 单元格2：保存错题到文本文件
# ============================================

def save_mistakes(mistake_list, filename="mistakes.txt"):
    """
    将错题列表保存到文本文件
    
    参数:
        mistake_list: 错题列表，格式 [{"question": "题目", "user_answer": "用户答案", "correct_answer": "正确答案"}, ...]
        filename: 保存的文件名
    """
    # 情况1：没有错题
    if not mistake_list:
        with open(filename, 'w', encoding='utf-8') as f:
            f.write("🎉 太棒了！本次练习没有错题！ 🎉\n")
        print(f"✅ 没有错题，已记录到 {filename}")
        return
    
    # 情况2：有错题
    with open(filename, 'w', encoding='utf-8') as f:
        # 写入头部信息
        f.write(f"📝 错题记录（共{len(mistake_list)}题）\n")
        f.write("=" * 50 + "\n")
        f.write(f"生成时间：{datetime.now().strftime('%Y-%m-%d %H:%M:%S')}\n")
        f.write("=" * 50 + "\n\n")
        
        # 逐条写入错题
        for idx, mistake in enumerate(mistake_list, 1):
            f.write(f"{idx}. 题目：{mistake['question']}\n")
            f.write(f"   你的答案：{mistake['user_answer']}\n")
            f.write(f"   正确答案：{mistake['correct_answer']}\n")
            f.write("-" * 40 + "\n")
    
    print(f"✅ 错题已保存到 {filename}，共{len(mistake_list)}道错题")

# 测试这个函数
print("函数 save_mistakes 定义完成")
```

```python
# ============================================
# 单元格3：显示成绩统计报告
# ============================================

def show_report(score, total, mistake_list):
    """
    打印统计报告到屏幕
    
    参数:
        score: 总分（每题10分）
        total: 总题数
        mistake_list: 错题列表
    """
    correct_count = total - len(mistake_list)
    accuracy = (correct_count / total) * 100 if total > 0 else 0
    
    # 打印报告（美化输出）
    print("\n" + "="*60)
    print("📊 成 绩 报 告".center(58))
    print("="*60)
    print(f"📝 总题数：{total}")
    print(f"✅ 正确数量：{correct_count}")
    print(f"❌ 错误数量：{len(mistake_list)}")
    print(f"🎯 总分：{score} / {total*10}")
    print(f"📈 正确率：{accuracy:.1f}%")
    print("="*60)
    
    # 根据正确率给出评价
    if accuracy == 100:
        print("🏆 完美！你是地理小天才！")
    elif accuracy >= 80:
        print("🌟 不错！继续努力！")
    elif accuracy >= 60:
        print("📚 加油！多练习几次！")
    else:
        print("💪 别灰心，多看看错题本！")
    
    print("="*60)
    
    # 询问是否查看详细错题
    if mistake_list:
        choice = input("\n📖 是否查看详细错题？(y/n): ").strip().lower()
        if choice == 'y':
            print("\n" + "="*60)
            print("错 题 详 情".center(58))
            print("="*60)
            for idx, mistake in enumerate(mistake_list, 1):
                print(f"\n{idx}. {mistake['question']}")
                print(f"   ❌ 你的答案：{mistake['user_answer']}")
                print(f"   ✅ 正确答案：{mistake['correct_answer']}")
            print("\n" + "="*60)
    else:
        print("\n🎉 恭喜！本次练习无错题！")
    
    return accuracy

print("函数 show_report 定义完成")
```

```python
# ============================================
# 单元格5：创新功能1 - 智能错题本（带历史记录）
# ============================================

import json
import csv
import os

def save_mistakes_advanced(mistakes_list, filename, formats=["txt"]):
    """
    将错题列表保存为多种格式的文件。
    """
    # 检查列表是否为空
    if not mistakes_list:
        print("📭 目前没有错题需要保存。")
        return

    print(f"💾 正在保存错题记录到 {filename}...")

    # 1. 保存为 TXT (纯文本)
    if "txt" in formats:
        txt_path = f"{filename}.txt"
        with open(txt_path, 'w', encoding='utf-8') as f:
            f.write("=== 错题本 ===\n\n")
            for i, item in enumerate(mistakes_list, 1):
                f.write(f"题目 {i}: {item['question']}\n")
                f.write(f"你的答案: {item['user_answer']}\n")
                f.write(f"正确答案: {item['correct_answer']}\n")
                f.write("-" * 30 + "\n")
        print(f"  ✅ 已生成: {txt_path}")

    # 2. 保存为 CSV (表格)
    if "csv" in formats:
        csv_path = f"{filename}.csv"
        with open(csv_path, 'w', newline='', encoding='utf-8-sig') as f:
            writer = csv.DictWriter(f, fieldnames=["question", "user_answer", "correct_answer"])
            writer.writeheader()
            writer.writerows(mistakes_list)
        print(f"  ✅ 已生成: {csv_path}")

    # 3. 保存为 HTML (网页)
    if "html" in formats:
        html_path = f"{filename}.html"
        html_content = "<!DOCTYPE html><html><head><meta charset='UTF-8'><title>错题本</title>"
        html_content += "<style>body{font-family:sans-serif;} table{border-collapse:collapse; width:100%;} th,td{border:1px solid #ddd; padding:8px;} th{background-color:#f2f2f2;} .wrong{color:red;}</style></head><body>"
        html_content += "<h2>我的错题本</h2><table><tr><th>题目</th><th>我的答案</th><th>正确答案</th></tr>"
        
        for item in mistakes_list:
            html_content += f"<tr><td>{item['question']}</td><td class='wrong'>{item['user_answer']}</td><td>{item['correct_answer']}</td></tr>"
            
        html_content += "</table></body></html>"
        
        with open(html_path, 'w', encoding='utf-8') as f:
            f.write(html_content)
        print(f"  ✅ 已生成: {html_path}")


class SmartMistakeBook:
    """智能错题本，记录错题出现次数，自动标记高频错题"""
    
    def __init__(self, filename="learning_report.json"):
        self.filename = filename
        self.data = self.load_history()
        print(f"📚 智能错题本已初始化，数据文件：{filename}")
    
    def load_history(self):
        """加载历史错题数据"""
        if os.path.exists(self.filename):
            with open(self.filename, 'r', encoding='utf-8') as f:
                return json.load(f)
        # 初始数据结构
        return {
            "mistakes": {},      # 存储错题详情
            "total_attempts": 0, # 总练习次数
            "total_mistakes": 0   # 总错题数量
        }
    
    def save_history(self):
        """保存历史数据到文件"""
        with open(self.filename, 'w', encoding='utf-8') as f:
            json.dump(self.data, f, ensure_ascii=False, indent=2)
        print(f"💾 数据已保存到 {self.filename}")
    
    def add_mistakes(self, mistake_list):
        """添加本次错题记录"""
        if not mistake_list:
            print("本次没有错题，太棒了！")
            self.data["total_attempts"] += 1
            self.save_history()
            return
        
        self.data["total_attempts"] += 1
        for mistake in mistake_list:
            question = mistake['question']
            wrong_answer = mistake['user_answer']
            # 创建唯一标识
            key = f"{question}|{wrong_answer}"
            
            if key not in self.data["mistakes"]:
                # 首次出现的错题
                self.data["mistakes"][key] = {
                    "question": question,
                    "wrong_answer": wrong_answer,
                    "correct_answer": mistake['correct_answer'],
                    "count": 1,
                    "first_occurrence": datetime.now().isoformat(),
                    "last_occurrence": datetime.now().isoformat()
                }
            else:
                # 重复错误的题目
                self.data["mistakes"][key]["count"] += 1
                self.data["mistakes"][key]["last_occurrence"] = datetime.now().isoformat()
            
            self.data["total_mistakes"] += 1
    
    def get_frequent_mistakes(self, threshold=2):
        """获取高频错题（重复错误超过threshold次）"""
        frequent = []
        for key, info in self.data["mistakes"].items():
            if info["count"] >= threshold:
                frequent.append(info)
        # 按错误次数从高到低排序
        frequent.sort(key=lambda x: x["count"], reverse=True)
        return frequent
    
    def show_learning_analytics(self):
        """显示学习分析报告"""
        print("\n" + "="*60)
        print("📈 学 习 分 析 报 告".center(58))
        print("="*60)
        print(f"📊 累计练习次数：{self.data['total_attempts']}")
        print(f"❌ 累计错题总数：{self.data['total_mistakes']}")
        print(f"📚 独特错题数量：{len(self.data['mistakes'])}")
        
        frequent = self.get_frequent_mistakes(threshold=2)
        if frequent:
            print(f"\n⚠️  高频错题（重复错误≥2次）：{len(frequent)}个")
            print("-"*60)
            for idx, mistake in enumerate(frequent[:5], 1):  # 只显示前5个
                print(f"{idx}. {mistake['question']}")
                print(f"   错误答案：{mistake['wrong_answer']} (出现{mistake['count']}次)")
                print(f"   正确答案：{mistake['correct_answer']}")
        else:
            print("\n🎉 没有高频错题，每次错误都在进步！")
        
        # 显示学习建议
        if self.data["total_attempts"] >= 3:
            total_unique = len(self.data['mistakes'])
            if total_unique > 10:
                print("\n💡 建议：错题较多，建议先复习基础知识")
            elif total_unique > 5:
                print("\n💡 建议：整理错题本，重点复习高频错题")
            else:
                print("\n💡 建议：保持练习，你已经很棒了！")
        
        print("="*60)

# 测试智能错题本
print("🔍 测试智能错题本...")
smart_book = SmartMistakeBook("my_learning_report.json")
smart_book.add_mistakes(test_mistakes)  # 添加第一次错题
smart_book.add_mistakes(test_mistakes)  # 模拟第二次又错了同样的题
smart_book.show_learning_analytics()
```

```python
# ============================================
# 单元格7：创新功能3 - 生成复习卡片
# ============================================

def generate_review_cards(mistake_list, filename="review_cards.txt"):
    """生成错题复习卡片（可打印剪裁）"""
    if not mistake_list:
        print("没有错题，无需生成复习卡")
        return
    
    with open(filename, 'w', encoding='utf-8') as f:
        f.write("="*60 + "\n")
        f.write("📇 错题复习卡（打印后沿虚线剪裁）\n")
        f.write("="*60 + "\n")
        f.write(f"生成时间：{datetime.now().strftime('%Y-%m-%d %H:%M:%S')}\n")
        f.write("="*60 + "\n\n")
        
        for idx, mistake in enumerate(mistake_list, 1):
            # 制作卡片边框
            f.write(f"┌{'─'*56}┐\n")
            f.write(f"│ 【复习卡 {idx:02d}】{' ':<46}│\n")
            f.write(f"├{'─'*56}┤\n")
            
            # 题目（可能换行）
            question = mistake['question']
            if len(question) <= 46:
                f.write(f"│ 📖 {question:<54}│\n")
            else:
                f.write(f"│ 📖 {question[:46]}{' ':<3}│\n")
                f.write(f"│   {question[46:92]:<54}│\n")
            
            f.write(f"├{'─'*56}┤\n")
            f.write(f"│ ❌ 你的答案：{mistake['user_answer']:<44}│\n")
            f.write(f"│ ✅ 正确答案：{mistake['correct_answer']:<44}│\n")
            f.write(f"├{'─'*56}┤\n")
            f.write(f"│ 【复习次数】第1次{' ':<45}│\n")
            f.write(f"│ 【掌握程度】□ 未掌握 □ 基本掌握 □ 已掌握{' ':<22}│\n")
            f.write(f"└{'─'*56}┘\n\n")
            f.write(f"{' ':<28}✂️ 裁剪线 ✂️{' ':<28}\n\n")
    
    print(f"✅ 复习卡片已生成：{filename}")
    print(f"   💡 提示：打印后剪裁，可随身携带复习")

def generate_flashcards_json(mistake_list, filename="flashcards.json"):
    """生成JSON格式的闪卡（可用于Anki等软件）"""
    if not mistake_list:
        print("没有错题，无法生成闪卡")
        return
    
    flashcards = []
    for mistake in mistake_list:
        flashcards.append({
            "front": mistake['question'],
            "back": f"正确答案：{mistake['correct_answer']}\n你的错误答案：{mistake['user_answer']}",
            "tags": ["地理", "错题", "中国省份"]
        })
    
    with open(filename, 'w', encoding='utf-8') as f:
        json.dump(flashcards, f, ensure_ascii=False, indent=2)
    
    print(f"✅ 闪卡数据已生成：{filename}")
    print(f"   💡 提示：此文件可导入 Anki 等记忆软件")

# 测试复习卡片
print("🔍 测试复习卡片生成...")
generate_review_cards(test_mistakes, "my_review_cards.txt")
generate_flashcards_json(test_mistakes, "my_flashcards.json")
```
